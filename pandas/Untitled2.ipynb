```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
# 함수 자체가 출력하는 함수이기 때문에 print()를 안 써도 됨.
# titanic.info() # 이 함수는 none을 리턴하기 때문에 print를 적으면 안 된다.

# dir(help) # 이 함수는 출력하지 않지만 jupyter notebook은 제일 마지막에 적으면 출력해줌.
help(titanic.info)
```

```python
df = titanic[['age', 'sex', 'class', 'fare', 'survived']]
print(df.head())
```

```python
grouped = df.groupby(['class']) # iterator의 구조를 가짐.
print(grouped)
dir(grouped)  # '__iter__'가 있으면 for을 쓸 수 있다.
```

```python
for key, group in grouped:
    print(key, group) # ['class']으로 grouped된 데이터.
```

```python
# 그룹 별 집계 가능
print(grouped.mean())
```

```python
# 활성화 함수 - 시그모이드
import math
import tensorflow as tf
import numpy as np
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# 입력 값
X = 1

# 실제 타겟
y = 0

# 가중치를 랜덤하게 생성
w = tf.random.normal([1], 0, 1)

output = sigmoid(X * w)
print(output)
```

```python
# Gradient Descent
for i in range(1000):
    output = sigmoid(X * w)
    error = y - output
    w = w + X * 0.1 * error
    if i % 50 == 0:
        print(i, output, error)
```

```python
# 경사 하강법 문제 발생
# 가중치만을 곱하는 경우 input이 0이면 가중치가 변경되지 않음
X = 0
y = 1

w = tf.random.normal([1], 0, 1)
b = tf.random.normal([1], 0, 1)
```

```python
# Gradient Descent
for i in range(1000):
    output = sigmoid(X * w + 1 * b)
    error = y - output
    w = w + X * 0.1 * error
    b = b + 1 * 0.1 * error
    if i % 50 == 0:
        print(i, output, error)
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [0], [0], [1]])
b = tf.random.normal([1], 0, 1)

#입력이 2개라서
w = tf.random.normal([2], 0, 1)

for i in range(2000):
    # 오차의 합계를 저장할 변수
    error_sum = 0

    # 모든 입력을 대입해서 수행
    for j in range(4):
        output = sigmoid(np.sum(X[j] * w) + 1 * b)
        error = y[j][0] - output
        w = w + X[j] * 0.1 * error
        b = b + 1 * 0.1 * error
        error_sum += error
    if i % 100 == 0:
        print(i, error_sum)
```

```python
# predict
for i in range(4):
    print('X:', X[i], 'Y:', y[i], 'Predict:', sigmoid(np.sum(X[i] * w) + b))
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [1]])
b = tf.random.normal([1], 0, 1)

#입력이 2개라서
w = tf.random.normal([2], 0, 1)

for i in range(2000):
    # 오차의 합계를 저장할 변수
    error_sum = 0

    # 모든 입력을 대입해서 수행
    for j in range(4):
        output = sigmoid(np.sum(X[j] * w) + 1 * b)
        error = y[j][0] - output
        w = w + X[j] * 0.1 * error
        b = b + 1 * 0.1 * error
        error_sum += error
    if i % 100 == 0:
        print(i, error_sum)
```

```python
# predict
for i in range(4):
    print('X:', X[i], 'Y:', y[i], 'Predict:', sigmoid(np.sum(X[i] * w) + b))
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=2, activation='sigmoid', input_shape=(2, )),
    tf.keras.layers.Dense(units=1, activation='sigmoid')])

model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1), loss='mse')

model.summary()

history = model.fit(X, y, epochs=2000, batch_size=1)

model.predict(X)

```

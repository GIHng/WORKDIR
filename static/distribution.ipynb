```python
import seaborn as sns
import scipy as sp
import scipy.stats

import numpy as np
import pandas as pd

# import pandas_datareader.data as web
#시각화 패키지
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager, rc

if platform.system() == 'Darwin': rc('font', family='AppleGothic')

import matplotlib 
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 정규 분포 만들기
# 평균과 표준편차가 필요함.

mu = 0 # 평균
std = 1 # 표준편차

rv = sp.stats.norm(mu, std)

xx = np.linspace(-5, 5, 10000000) # 샘플 데이터 100개를 생성

plt.plot(xx, rv.pdf(xx))
plt.title('평균이 0이고 표준 편차가 1인 정규 분포')
plt.show()
```

```python
x = rv.rvs(100)
print(x)

plt.plot(xx, rv.cdf(xx))
```

```python
mu = 1
rv = sp.stats.norm(loc=mu)

x1 = rv.rvs(1000000)
s = 0.5
x2 = np.exp(s * x1)

fig, ax = plt.subplots(1, 2)
sns.distplot(x1, kde=False, ax= ax[0])
sns.distplot(x2, kde=False, ax= ax[1])
plt.show()


# 로그 정규 분포 모형이 1에 수렴하는 이유.
# 로그 정규 분포는 항상 양수 값만을 가짐.
# 주식같은 경우 일정한 비율이 곱해지며 변화하는 성질이 있기 때문에, 로그 정규 분포를 띌 것으로 예상함.
```

```python
data_poisson = sp.stats.poisson.rvs(mu=3, size= 100000)
print(data_poisson)

ax = sns.distplot(data_poisson, bins=30, color='blue', kde=False)
plt.show()
```

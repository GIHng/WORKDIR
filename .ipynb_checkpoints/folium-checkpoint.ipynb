```python
import numpy as np
import pandas as pd
import platform
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import folium


m = folium.Map(location = [37.572656, 126.973304], zoom_start = 15)

# Marker 출력
folium.Marker(location=[36.572656, 126.973304], popup="KB카드", icon=folium.Icon(icon='cloud')).add_to(m)
folium.Marker(location=[36.569027, 126.987029], popup="Mega IT", icon=folium.Icon(icon='cloud')).add_to(m)

folium.RegularPolygonMarker(location=[37.569027, 126.987029], popup="PolygonMarker",number_of_sides=5, rotation=0, radius=60, fill = True, fill_color="#999999").add_to(m)

m
```

```python
df = pd.read_excel('./data/서울지역_대학교_위치.xlsx')
map = folium.Map(location = [37.55, 126.98], zoom_start = 12)

for name, lat, lon in zip(df['Unnamed: 0'], df['위도'], df['경도']):
    folium.CircleMarker([lat, lon],popup=name, radius=30, color='blue', fill=True, fill_color='coral', fill_opacity=0.7).add_to(map)

map
```

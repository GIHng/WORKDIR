```python
import numpy as np
import pandas as pd
import platform
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# matplotlib 기반의 그래프를 출력 시 한글 사용을 위한 설정

# system이 Mac이면,
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
    
# system이 Windows이면,
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(
    fname='c:/Windows/Fonts/malgun.ttf').get_name()
    rc('font', family=font_name)
```

```python
df = pd.read_csv('./data/student.csv', encoding='ms949')

plt.figure()
colormap = df.index
plt.scatter(x = df.index, y=df['국어'], marker = '2', c=colormap)
plt.title('학생별 국어 점수 분포')
plt.xticks(range(0, len(df['국어']), 1), df['이름'], rotation='vertical')
plt.colorbar()
plt.show()
```

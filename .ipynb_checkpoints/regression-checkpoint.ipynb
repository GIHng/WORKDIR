```python
import numpy as np
import pandas as pd
import platform
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# matplotlib 기반의 그래프를 출력 시 한글 사용을 위한 설정

# system이 Mac이면,
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
    
# system이 Windows이면,
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(
    fname='c:/Windows/Fonts/malgun.ttf').get_name()
    rc('font', family=font_name)
```

```python
# noheader_auto-mpg.csv파일을 불러 옴
df = pd.read_csv('./data/noheader_auto-mpg.csv', header=None)
df.columns=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
# print(df.head())
df['cylinders_size'] = df['cylinders'] / df.cylinders.max() * 300
print(df.head())

# 회귀 분석에 사용 (다항 회귀)
```

```python
# 그래프
plt.scatter(x=df['weight'], y=df['mpg'], s=df['cylinders_size'], c='coral', alpha=0.2)
plt.title('중량과 연비의 관계')
plt.show()
```

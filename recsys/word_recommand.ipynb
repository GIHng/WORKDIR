```python
from bs4 import BeautifulSoup
```

```python
import urllib
import time
from tqdm import tqdm_notebook
```

```python
present_candi_text = []

html = 'https://search.naver.com/search.naver?where=kin&sm=tab_jum&ie=utf8&query=' + urllib.parse.quote('당구') + '&kin_start='
for n in range(1, 500, 10):
    print(html+str(n))
    response = urllib.request.urlopen(html + str(n))

    soup = BeautifulSoup(response, 'html.parser')
    tmp = soup.select('div.question_area > div.question_group > a')
    for line in tmp:
        print(line.text)
        present_candi_text.append(line.text)
    time.sleep(1)
print(present_candi_text)
```

```python
import nltk
from konlpy.tag import Okt
okt = Okt()
present_text = ''

for each_line in present_candi_text[:10000]:
    present_text = present_text + each_line + '\n'
tokens_ko = okt.morphs(present_text)
print(tokens_ko)
```

```python
ko = nltk.Text(tokens_ko, name='당구')
print(ko.vocab().most_common(100))
```

```python
# stopwords
stop_words = ['...', '..', '.', '. ', '\n', '?', ]
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
tokens_ko = [each_word for each_word in tokens_ko if len(each_word) > 1]


ko = nltk.Text(tokens_ko, name='당구')
ko.vocab().most_common(50)
```

```python
import matplotlib.pyplot as plt
```
